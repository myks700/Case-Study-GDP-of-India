```python
states=['Andhra_Pradesh',
 'Arunachal_Pradesh',
 'Assam',
 'Bihar',
 'Chhattisgarh',
 'Goa',
 'Gujarat',
 'Haryana',
 'Himachal_Pradesh',
 'Jammu & Kashmir',
 'Jharkhand',
 'Karnataka',
 'Kerala',
 'Madhya_Pradesh',
 'Maharashtra',
 'Manipur',
 'Meghalaya',
 'Mizoram',
 'Nagaland',
 'Odisha',
 'Punjab',
 'Rajasthan',
 'Sikkim',
 'Tamil Nadu',
 'Telangana',
 'Tripura',
 'Uttar_Pradesh',
 'Uttarakhand',
 'West Bengal']
```

```python
import os
```

```python
gsva=pd.DataFrame()
```

```python
path=("/home/mayank/Case Study/GDP Case Study/Data Set/GSVA/")
```

```python
listOfFiles=os.listdir(path)
```

```python
len(listOfFiles)
```

```python
for filename in listOfFiles:
    for statename in states:
        if statename in filename:
            filepath=path+filename
            data=pd.read_csv(filepath,encoding="ISO-8859-1")
            data=data[['Item','2014-15']]
            data=data.T
            data.columns=data.iloc[0]
            data=data[1:]
            data["state"]=statename
            gsva=pd.concat([data,gsva])
```

```python
gsva.set_index("state", inplace=True)
```

```python
category_1 = ["Goa", "Sikkim", "Haryana", "Kerala"]
category_2 = ["Arunachal_Pradesh", "Punjab", "Telangana", "Gujarat", "Karnataka","Maharashtra", "Uttarakhand"]
category_3 = ["Madhya_Pradesh", "Odisha", "Meghalaya", "Tripura", "Rajasthan", "Chhattisgarh","Nagaland","Andhra_Pradesh"]
category_4 = ["Bihar", "Uttar_Pradesh", "Manipur", "Assam", "Jharkhand"]

categories_columns = ['Agriculture, forestry and fishing','Mining and quarrying', 'Manufacturing',
       'Electricity, gas, water supply & other utility services',
       'Construction', 'Trade, repair, hotels and restaurants',
       'Transport, storage, communication & services related to broadcasting',
       'Financial services','Real estate, ownership of dwelling & professional services',
       'Public administration', 'Other services', 'Taxes on Products','Subsidies on products']
```

```python
gsva
```

## Category 1 States

```python
agg_1=gsva.loc[category_1,categories_columns].sum()
agg_gsdp_1=gsva.loc[category_1,"Gross State Domestic Product"].sum()
Category_1=round(agg_1/agg_gsdp_1*100).sort_values(ascending=False).head(6)
Category_1.plot.bar()
plt.show()
```

**Category 1**: There are only `4 states` in this category, all of which are small (area-wise) and have good industrial base , agricultural, trade and real estate.

* **Trade and hospitality** business should be promoted further
* **Manufacturing** should be further promoted

```python
agg_1.index
```

## Category 2 States 

```python
agg_2=gsva.loc[category_2,categories_columns].sum()
agg_gsdp_2=gsva.loc[category_2,"Gross State Domestic Product"].sum()
Category_2=round(agg_2/agg_gsdp_2*100).sort_values(ascending=False).head(6)
Category_2.plot.bar()
plt.show()
```

* These states have relatively `good per capita gdp` and are driven by around equal percentage by industries and agriclture. These implies that these states have pool of talent to drive industries and agriculture. Because of the talent, other industries can take their advantage and set up good industries.
* States like `karnataka, tamil nadu , gujarat and maharashtra` which has huge coastal area can setup ports. Coastal area can help in harness clean energy and fisheries can be increased. Similarly, tourism industry should also be encouraged.

## Category 3 States

```python
agg_3=gsva.loc[category_3,categories_columns].sum()
agg_gsdp_3=gsva.loc[category_3,"Gross State Domestic Product"].sum()
Category_3=round(agg_3/agg_gsdp_3*100).sort_values(ascending=False).head(6)
Category_3.plot.bar()
plt.show()
```

**Madhya_Pradesh, Odisha, Meghalaya, Tripura, Rajasthan,Chhattisgarh, Nagaland, Andhra_Pradesh are included in category 3**
* Have a good base of Agricultural, Forestry and Fishing.
* Real estate, ownership of dwelling & professional services and Construction Industries need to be consider more here.

## Category 4 States

```python
agg_4=gsva.loc[category_4,categories_columns].sum()
agg_gsdp_4=gsva.loc[category_4,"Gross State Domestic Product"].sum()
Category_4=round(agg_4/agg_gsdp_4*100).sort_values(ascending=False).head(6)
Category_4.plot.bar()
plt.show()
```

**Bihar, Uttar_Pradesh, Manipur, Assam, Jharkhand** are included in ths category
* Agriculture, forestry and fishing have good base here.
* Contruction need to be more consider here

```python
fig=plt.figure()


ax = fig.add_axes([-.5, 0, .55, .8], aspect=1)
ax.pie(Category_1,labels=Category_1.index, radius= 1.2, autopct="%.2f%%", startangle=90)
ax.set_title("Category 1")

ax1 = fig.add_axes([.8, .0, .55, .8], aspect=1)
ax1.pie(Category_2,labels=Category_2.index, radius= 1.2, autopct="%.2f%%", startangle=90)
ax1.set_title("Category 2")

fig=plt.figure()

ax2 = fig.add_axes([-.5, 0, .55, .8], aspect=1)
ax2.pie(Category_3,labels=Category_3.index, radius= 1.2, autopct="%.2f%%", startangle=90)
ax2.set_title("Category 3")

ax3 = fig.add_axes([1, 0, .55, .8], aspect=1)
ax3.pie(Category_4,labels=Category_4.index, radius= 1.2, autopct="%.2f%%", startangle=90)
ax3.set_title("Category 4")

plt.show()
```
